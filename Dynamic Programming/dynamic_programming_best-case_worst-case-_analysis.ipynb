```python
import time
import random
import matplotlib.pyplot as plt
import job_scheduling_dynamic_programming as algorithm
```

```python
def generateJobs(n):
    # Generate n number of random jobs with random deadlines and profits
    jobs = []
    for i in range(n):
        jobId = i + 1
        deadline = random.randint(1, 200)
        profit = random.randint(0, 500)
        jobs.append((jobId, deadline, profit))
    return jobs
```

```python
def measureExecutionTime(jobs):
    start_time = time.time()
    # Call the schedule_jobs function with the given jobs
    algorithm.jobSchedulingWithDynamicProgramming(jobs)
    end_time = time.time()
    execution_time = end_time - start_time
    return execution_time
```

```python
def plotTimeComplexity(generatedValues, bestCase, worstCase):
    plt.plot(generatedValues, bestCase, label='Best Case')
    plt.plot(generatedValues, worstCase, label='Worst Case')
    plt.xlabel('Number of Jobs')
    plt.ylabel('Execution Time')
    plt.title('Time Complexity Analysis of Dynamic Programming')
    plt.legend()
    plt.show()
```

```python
jobGenerator = range(50, 500, 10)  # 10 to 100 with a step size of 100

bestCaseTimes = []
worstCaseTimes = []

# Measure execution time for each input size
for n in jobGenerator:
    jobs = generateJobs(n)

    # Sort jobs in descending order based on profits
    jobs.sort(key=algorithm.getJobProfit, reverse=True)

    # Best case: Jobs are already sorted in descending order
    bestCase = measureExecutionTime(jobs)
    bestCaseTimes.append(bestCase)

    # Worst case: Jobs are sorted in ascending order
    jobs.sort(key=algorithm.getJobProfit)
    worstCase = measureExecutionTime(jobs)
    worstCaseTimes.append(worstCase)
# Plot the time complexity graph
plotTimeComplexity(jobGenerator, bestCaseTimes, worstCaseTimes)
```
